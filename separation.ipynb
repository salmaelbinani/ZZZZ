```python
import re 
# Fonction pour extraire les chiffres de la colonne Level
def extract_level(level_str):
    match = re.search(r'\d+', level_str)
    return int(match.group()) if match else None

# Fonction pour mettre à jour la colonne PN
def update_pn(row):
    level = extract_level(row['Level'])
    if level == 0:
        return row['Material Description']
    return None

# Appliquer la fonction update_pn et remplir les NaN
df['PN'] = df.apply(update_pn, axis=1)
df['PN'] = df['PN'].fillna(method='ffill')
df_copie = df.copy()
# Afficher les premières lignes pour vérifier le résultat
df.head(20)
```

```python
# Fonction pour déterminer la "valeur" basée sur le nombre d'astérisques
def count_asterisks(value):
    return value.count('*')

# Appliquer cette fonction à la colonne 'Level'
df_1['asterisk_count'] = df_1['Level'].apply(count_asterisks)

# Trouver la valeur maximale en fonction du nombre d'astérisques
max_asterisks = df_1['asterisk_count'].max()
n=max_asterisks+1
#print(n)

# Calculer le nombre de nouvelles colonnes à créer
num_new_columns = (n - 1) * 2

# Créer les nouvelles colonnes
for i in range(1, n):
    df_1[f'SN{i}'] = None
    df_1[f'DSN{i}'] = None

# Supprimer la colonne intermédiaire utilisée pour le comptage des astérisques
df_1.drop(columns=['asterisk_count'], inplace=True)
df_1.head()
```

```python
# Define the function to update SN columns based on Level
def update_sn(row, level_indicator):
    if row['Level'] == level_indicator:
        return row['Material']

# General function to update and fill SN columns
def update_and_fill_sn(df_1, sn_column, level_indicator):
    df_1[sn_column] = df_1.apply(lambda row: update_sn(row, level_indicator), axis=1)
    df_1[sn_column] = df_1[sn_column].fillna(method='ffill')

# Fonction pour générer les niveaux
def generate_level(n):
    return '*' * (n + 1) + str(n + 1)

# Généraliser la boucle pour n colonnes
#n = 3  # Par exemple, 3 colonnes
for i in range(n-1):
    column_name = f'SN{i + 1}'
    level = generate_level(i)
    update_and_fill_sn(df_1, column_name, level)


df_1.head(15)
```

```python
# Create a dictionary mapping 'Material' to 'Material Description'
material_description_map = df_1.set_index('Material')['Material Description'].to_dict()

# List of SN and DSN columns to update
sn_columns = [f'SN{i}' for i in range(1, n)]
dsn_columns = [f'DSN{i}' for i in range(1, n)]
# General function to fill DSN columns based on SN columns
def fill_dsn_columns(df_1, sn_columns, dsn_columns, material_description_map):
    for sn_col, dsn_col in zip(sn_columns, dsn_columns):
        df_1[dsn_col] = df_1[sn_col].map(material_description_map)

# Fill the DSN columns
fill_dsn_columns(df_1, sn_columns, dsn_columns, material_description_map)

df_1.head(15)
```

```python
# Fonction pour nettoyer le tableau selon la valeur de 'Level'
def clean_table(row):
    if row['Level'] == '*1':
        row['SN2'] = '(blank)'
        row['DSN2'] = '(blank)'
        row['SN3'] = '(blank)'
        row['DSN3'] = '(blank)'
    elif row['Level'] == '**2':
        row['SN3'] = '(blank)'
        row['DSN3'] = '(blank)'
    return row

# Appliquer cette fonction à chaque ligne du DataFrame
df_1 = df_1.apply(clean_table, axis=1)

# Filtrer les lignes ayant au moins une valeur non manquante dans les colonnes SN2, DSN2, SN3, DSN3
filtered_df = df_1.dropna(subset=['SN2', 'DSN2', 'SN3', 'DSN3'], how='all')

# Garder uniquement les colonnes nécessaires pour la table pivotée
columns_to_keep = ['SN1', 'DSN1', 'SN2', 'DSN2', 'SN3', 'DSN3', 'PN']
filtered_df = filtered_df[columns_to_keep]

# Grouper et compter les occurrences
grouped = filtered_df.groupby(['SN1', 'DSN1', 'SN2', 'DSN2', 'SN3', 'DSN3', 'PN']).size().reset_index(name='count')

# Créer la table pivotée manuellement
pivot_table = grouped.pivot(index=['SN1', 'DSN1', 'SN2', 'DSN2', 'SN3', 'DSN3'], columns='PN', values='count').fillna(0).reset_index()

# Remplacer les zéros par NaN pour avoir des cellules vides dans le fichier Excel
pivot_table.replace(0, np.nan, inplace=True)

# Ajouter une colonne 'Total' qui somme les valeurs de chaque ligne
pivot_table['Total'] = pivot_table.sum(axis=1, numeric_only=True)


```

```python
import pandas as pd
from openpyxl import load_workbook

# Charger le fichier Excel et toutes ses feuilles
file_path = 'MMSTA_separe.xlsx'
excel_book = load_workbook(file_path)
sheet_names = excel_book.sheetnames

# Lire toutes les feuilles dans un dictionnaire
dfs = {sheet: pd.read_excel(file_path, sheet_name=sheet) for sheet in sheet_names}

# Ajouter la colonne "DS Général" après "SN1" dans la feuille "FIL SIMPLE"
df_fil_simple = dfs['FIL SIMPLE']
sn1_index = df_fil_simple.columns.get_loc("SN1")
df_fil_simple.insert(sn1_index + 1, "DS Général", "")

# Ajouter la colonne "DS Général" après "SN2" dans la feuille "joint"
df_joint = dfs['joint']
sn2_index = df_joint.columns.get_loc("SN2")
df_joint.insert(sn2_index + 1, "DS Général", "")

# Ajouter la colonne "DS Général" après "SN2" dans les feuilles "double", "twist", "SQUIB" et "GW"
for sheet in ['double', 'twist', 'SQUIB', 'GW']:
    df = dfs[sheet]
    sn2_index = df.columns.get_loc("SN2")
    df.insert(sn2_index + 1, "DS Général", "")
    dfs[sheet] = df

# Ajouter la colonne "DS Général" après "SN3" dans la feuille "super group"
df_super_group = dfs['super group']
sn3_index = df_super_group.columns.get_loc("SN3")
df_super_group.insert(sn3_index + 1, "DS Général", "")

# Fonction pour générer "DS Général" pour une valeur donnée de "SN1", "SN2" ou "SN3"
def generate_ds_general(sn_value):
    if pd.isna(sn_value):
        return ''
    
    # Trouver la ligne de départ pour la valeur SN dans le DataFrame df_mmsta
    start_rows = df_mmsta[df_mmsta['Material'] == sn_value]
    
    if start_rows.empty:
        return ''
    
    start_index = start_rows.index[0]
    ds_general_parts = []
    
    # Concaténer les valeurs verticalement à partir de la ligne de départ
    for i in range(start_index + 1, len(df_mmsta)):
        cell_value = df_mmsta.at[i, 'Material']
        if pd.isna(cell_value):
            continue
        cell_value_str = str(cell_value)
        if cell_value_str.startswith('S00'):
            break
        if df_mmsta.at[i, 'Material Description'].startswith('LT Single Wire'):
            component_quantity = df_mmsta.at[i, 'Component quantity']
            ds_general_parts.append(f"{cell_value_str} ({component_quantity})")
        else:
            ds_general_parts.append(cell_value_str)
    
    return ' / '.join(ds_general_parts)

# Appliquer la fonction à chaque valeur de "SN1" dans "FIL SIMPLE" et remplir "DS Général"
df_mmsta = dfs['MMSTA']
df_fil_simple["DS Général"] = df_fil_simple["SN1"].apply(generate_ds_general)

# Appliquer la fonction à chaque valeur de "SN2" dans "joint", "double", "twist", "SQUIB", et "GW" et remplir "DS Général"
for sheet in ['joint', 'double', 'twist', 'SQUIB', 'GW']:
    df = dfs[sheet]
    df["DS Général"] = df["SN2"].apply(generate_ds_general)
    dfs[sheet] = df

# Appliquer la fonction à chaque valeur de "SN3" dans "super group" et remplir "DS Général"
df_super_group["DS Général"] = df_super_group["SN3"].apply(generate_ds_general)

# Sauvegarder les modifications dans le fichier Excel original
with pd.ExcelWriter(file_path, engine='openpyxl') as writer:
    for sheet_name, df in dfs.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)

print(f"Le fichier a été mis à jour et sauvegardé sous '{file_path}'")
```
